```python
x['Invoice Date'] = x['Invoice Date'].apply(lambda my_date:parse(my_date).strftime('%d/%m/%y'))
```

```python
#changing the column name by using rename()
#inpalce=true means to modify the original dataset directiy, without creating new copy. if you donot keep it will not modiify the origin 
x.rename(columns={'Retailer ID':'Retailer_ID','Invoice Date':'Invoice_Date','Price per Unit':'Price_per_unit','Units Sold':'Units_sold','Total Sales':'Total_sales','Operating Profit':'Operating_profit','Operating Margin':'Operating_margin','Sales Method':'Sales_method'},inplace=True)
```

```python
#REMOVEING THE DOLLAR SYMBOL

x[['Price_per_unit','Total_sales','Operating_profit']] = x[['Price_per_unit','Total_sales','Operating_profit']].replace({'\$':''},regex=True)
```

```python
x
```

```python
#strip()functions is bulit in string thatremove specified charcters from the beginng and end of string 
#lambda function is a small anonymous function that can be defined without a name and we donot want to define a separate functions 
x['Operating_margin']=x['Operating_margin'].apply(lambda my_data:float(my_data.strip('%'))/100)
```

```python
x
```

```python
x['Price_per_unit'] =x['Price_per_unit'].apply(lambda my_data:int(round(float(my_data))))
```

```python
# changing data type and removing comma
data_to_convert = ['Units_sold','Total_sales','Operating_profit']
for i in data_to_convert:
    x[i] = x[i].str.replace(',','').astype(int)
```

```python
sorted_data=x.sort_values("Total_sales",ascending=False)
Retailer=sorted_data['Retailer']
Total_sales=sorted_data['Total_sales']
plt.bar(Retailer,Total_sales,color='blue')
plt.xlabel('Retailer')
plt.ylabel('Total_sales')
plt.title('Sales by Retailer')
plt.xticks(rotation=45)
plt.show()
```

```python
sales_by_region= x.groupby('Region')['Total_sales'].sum()
sales_by_region=sales_by_region.sort_values(ascending=False)
```

```python
sales_by_region
```

```python
#analyze sales trends across differen region identify regions with the highest sales
sales_by_region= x.groupby('Region')['Total_sales'].sum()
sales_by_region=sales_by_region.sort_values(ascending=False)
fig,ax=plt.subplots(figsize=(8,10))#create a figure and axis
sales_by_region.plot(kind='bar',ax=ax,color='blue',alpha=0.7)
sales_by_region.plot(kind='line',ax=ax,marker='o',color='red')
ax.set_xlabel('Region')
ax.set_ylabel('Total_sales')
ax.set_title('Sales by Region with Trend line and Bar')
plt.show()
```
