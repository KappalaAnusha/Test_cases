3.Represent Corelation between variables ?

```python
data.corr(numeric_only=True)
```

```python
correlation_matrix=data.corr(numeric_only=True)
print(correlation_matrix)
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.title('correlation Matrix')
plt.show()
```

4.plot different kinds of graphs for mean value represetation ? (use required headers) ?

```python
mean_values=data.mean(numeric_only=True)
```

```python
mean_values
```

```python
mean_values.plot(kind='bar')
plt.xlabel('Variables')
plt.ylabel('Mean Values')
plt.title('Mean Values for Each Variables')
plt.show()
```

```python
mean_values.plot(kind='line')
plt.xlabel('Variables')
plt.ylabel('Mean Value')
plt.title('Mean Values for Each Variables over time')
plt.show()
```
