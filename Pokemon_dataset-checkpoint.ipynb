What are the most common types of Pokemon?

What are the strongest and weakest Pokemon species?

```python
strongest_pokemon=data.sort_values('Total',ascending=False)#find the strongest pokemon species by sorting total columns

weakest_pokemon=data.sort_values('Total')

print('strongest pokemon species:',strongest_pokemon[['Total','Name']])
print('weakest pokemon species:',weakest_pokemon[['Total','Name']])
```

```python
strongest_pokemon_4=data.loc[data['Total'].idxmax(),'Name']#find the strongest pokemon species by sorting total columns

weakest_pokemon_5=data.loc[data['Total'].idxmin(),'Name']

print('strongest pokemon species:',strongest_pokemon_4)
print('weakest pokemon species:',weakest_pokemon_5)
```

What are the strongest and weakest types of Pokemon?

```python
type_stats=data.groupby('Type_1')['Total'].mean()#group the pokemon by their typestats and calculate the avg total for each type

strongest_type=type_stats.idxmax()

weakest_type=type_stats.idxmin()

print('strongest type of pokemon:',strongest_type)
print('weakest type of pokeman:',weakest_type)
```

Do any types of Pokemon excel at certain statistics over others?
